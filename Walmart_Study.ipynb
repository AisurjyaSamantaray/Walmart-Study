```python
import scipy.stats as stats
from scipy.stats import ttest_ind
```

```python
#Perform an independent t-test to compare the sales before and after the campaign of a store
#Use a significance level (alpha) of 0.05.
#Report the p-value and interpret the result (whether to reject or fail to reject the null hypothesis).


#Null Hypothesis (H0): The promotion had no effect on the sales (i.e., the mean sales before and after the campaign are the same).
#Alternative Hypothesis (H1): The promotion had an effect on the sales (i.e., the mean sales after the campaign are different from before).

#Creating another dataframe to store a new feature ['campign period']

campign_data = data.copy(deep=True)
campign_data['Date'] = pd.to_datetime(campign_data['Date'], format='%d-%m-%Y')

#now creating a new feature for camign_data['camign_period']::assuming campign was launched on '02-03-2012'
cut_off_date = pd.to_datetime('02-03-2012')
campign_data['camign_period'] = ['Before' if i < cut_off_date else 'After' for i in campign_data['Date']]

sales_before_campaign = campign_data[campign_data['camign_period'] == 'Before']['Weekly_Sales']
sales_after_campaign = campign_data[campign_data['camign_period'] == 'After']['Weekly_Sales']

# Perform the T-test
t_stat, p_val = ttest_ind(sales_before_campaign, sales_after_campaign, equal_var=False)

print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

```python
# Group by 'Store' and 'Campaign_Period'
grouped = campign_data.groupby(['Store', 'camign_period'])['Weekly_Sales']

# Store-specific T-test
for store in campign_data['Store'].unique():
    # Get sales before and after the campaign for each store
    sales_before = grouped.get_group((store, 'Before'))
    sales_after = grouped.get_group((store, 'After'))

    # Perform the t-test if both groups have data
    if len(sales_before) > 1 and len(sales_after) > 1:
        t_stat, p_val = ttest_ind(sales_before, sales_after, equal_var=False)
        print(f"Store {store}: T-statistic = {t_stat}, P-value = {p_val}")
    else:
        print(f"Store {store}: Not enough data for t-test")
```

```python
#To test if there is a significant difference in the average weekly sales between different months across all stores.

#Null Hypothesis (H0): The average weekly sales are the same across all months.
#Alternative Hypothesis (H1): The average weekly sales are not the same across all months.
```

To Perform Annova we need to check the below things:
1. Check Normality of Distribution:
Histogram: To visualize the distribution.
Shapiro-Wilk Test: To statistically test normality.
Q-Q Plot: To visually inspect normality.
Levene's Test: To check the homogeneity of variances.
If these assumptions aren't met, I can consider a Kruskal-Wallis test as a non-parametric alternative to one-way ANOVA.
2. Perform One-Way ANOVA (if assumptions are satisfied):
Create a new random variable or grouping column (e.g., month or store).
Compare the means across groups using one-way ANOVA.

```python
sales = data['Weekly_Sales']
# Plot a histogram to visualize the distribution
sns.histplot(sales, kde=True)
plt.title("Distribution of Weekly Sales")
plt.show()

# Q-Q plot to visually inspect normality
from scipy.stats import shapiro, levene, probplot
probplot(sales, dist="norm", plot=plt)
plt.title("Q-Q Plot")
plt.show()

# Shapiro-Wilk test for normality
stat, p_value_shapiro = shapiro(sales)
print(f'Shapiro-Wilk Test: Statistic={stat}, P-value={p_value_shapiro}')
```

```python
#For levene
# Group by month
grouped_sales_by_month = data.groupby('Month')['Weekly_Sales'].apply(list)

# Perform Levene’s test by unpacking the grouped sales data
stat_levene, p_value_levene = levene(*grouped_sales_by_month)
print(f'Levene’s Test: Statistic={stat_levene}, P-value={p_value_levene}')
```

Levene’s Statistic: The statistic value (13.77) indicates the extent of variance difference between the groups. Higher values suggest greater differences in variance.

P-value: The very small p-value (1.61e-26) indicates that the variances of weekly sales across different months are significantly different from each other. In other words, the assumption of homogeneity of variances is violated.

Shapiro-Wilk Statistic: The statistic value (0.947) is a measure of how well the data matches a normal distribution. Values close to 1 suggest that the data is normally distributed.

P-value: The very small p-value (3.82e-43) suggests that the data significantly deviates from a normal distribution. In most cases, a p-value less than 0.05 indicates that we reject the null hypothesis of normality.

Still Assuming the condition are right and performing 1-way annova, to observe that the code works well.

```python
# Perform one-way ANOVA by passing the grouped sales data
from scipy.stats import f_oneway

stat_anova, p_value_anova = f_oneway(*grouped_sales_by_month)
print(f"ANOVA Test: Statistic={stat_anova}, P-value={p_value_anova}")
```

```python
from scipy.stats import kruskal

# Perform Kruskal-Wallis test (non-parametric alternative to one-way ANOVA)
stat_kruskal, p_value_kruskal = kruskal(*grouped_sales_by_month)
print(f"Kruskal-Wallis Test: Statistic={stat_kruskal}, P-value={p_value_kruskal}")
```

Kruskal-Wallis Statistic: The statistic value (61.92) reflects the magnitude of differences between the medians of the groups. Higher values indicate more substantial differences.

P-value: The very small p-value (4.08e-09) suggests that there are significant differences between the medians of weekly sales across the different months.

Action Based on Results:
Since the p-value is very small, I should reject the null hypothesis that all groups have the same median. This indicates that there are statistically significant differences in the medians of weekly sales across different months.

Summary of Hypothesis Testing:
Shapiro-Wilk Test: Data is not normally distributed.
Levene’s Test: Variances are not homogeneous across groups.
Kruskal-Wallis Test: Significant differences in medians across months.
This analysis suggests that the sales data across months differ significantly, although the assumptions for parametric tests (like ANOVA) were not met.
